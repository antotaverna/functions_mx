Simulated data

```python
# Matrix similar to real data
lvx=8
matrix = np.random.randint(1, 5, size=(lvx, lvx, lvx))

print(matrix.shape)
print(np.unique(matrix))
```

```python
xmin=0; xmax=lvx
ymin=0; ymax=lvx
zmin=0; zmax=lvx



def mask_box(_x,_y,_z):
    idx_mask = np.where((_x>xmin)&(_x<xmax)& (_y>ymin)&(_y<ymax) & (_z>zmin)&(_z<zmax))
    _xx=_x[idx_mask]
    _yy=_y[idx_mask]
    _zz=_z[idx_mask]
    
    return np.array(_xx), np.array(_yy), np.array(_zz)


xg, yg, zg = np.where(matrix==1)
x1, y1, z1 = mask_box(xg, yg, zg)
print(np.shape(x1))

xfof, yfof, zfof = np.where(matrix==2)
x2, y2, z2 = mask_box(xfof, yfof, zfof)
print(np.shape(x2))


data1 = plot_go_3d(x1, y1, z1, col='blue', op=0.2,sz=3, name='Data1')
data2 = plot_go_3d(x2, y2, z2, col='red', op=0.7,sz=3, name='Data2')


fig = go.Figure(data=[data1, data2])

fig.update_layout(width=500, height=500)
fig.show()
```

```python

data1_vx = plot_mesh(x1, y1, z1, col='blue', op=0.5, voxel_size=0.8, name='Data1')
data2_vx = plot_mesh(x2, y2, z2, col='red', op=0.5, voxel_size=0.5, name='Data2')


fig = go.Figure(data=[data1_vx, data2_vx])

fig.update_layout(width=500, height=500)
fig.show()
```
